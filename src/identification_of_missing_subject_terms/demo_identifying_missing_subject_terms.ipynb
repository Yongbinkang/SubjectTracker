# Demonstrate identifying missing subject terms program

- Use sample description and apoSubject data to identify missing subject terms. 
- Identifying missing subject terms program process as follows:
    1. Read apoDescriptions and apoSubject data files
    2. Extract text from apoDescription data
    3. Divide subject terms into uppercase and lowercase
    4. Apply regular expressions to search uppercase and lowercase subject terms in text respectively

**Set parameters in params/param.dat**

- DESCRIPTION_DATA: directory of description data (e.g. ../data/apoDescriptions.csv)
- SUBJECT_TERM_DATA: directory of subject data (e.g. ../data/apoSubjects.csv)
- OUTPUT_DIR : directory of output data files (e.g. ../data)

```python
# param.dat

DESCRIPTION_DATA = '../../data/apoDescriptions.csv'
SUBJECT_TERM_DATA = '../../data/apoSubjects.csv'
OUTPUT_DIR = '../../data'
```
