# Demonstrate inducing a subject term taxonomy program

- Use sample data to demonstrate how to induce a subject term taxonomy with refined subject terms.
- Inducing subject term taxonomy program process as follows:
    1. Read merged_subject_term.csv (result of the filtering out irrelevant subject terms)
    2. Replace secondary subject term to primary subject term based on their similarity
    3. Induce a subject term with subsumption method

## Set parameters in params/param.dat

- ROOT: Define name of ROOT Node(e.g. ROOT)
- DATA_PATH: directory of output data (e.g. ../sample_data/)
- THRESHOLD: threshold value to build taxonomy (e.g. 0.2)
- MIN_DF: integer to set minimal document frequency for subject term

## Visualize generated subject term taxonomy

```python
import pickle
import networkx as nx
import matplotlib.pyplot as plt
import networkx as nx
import pydot
from networkx.drawing.nx_pydot import graphviz_layout

labels = {}
path_taxonomy = "../../data/taxonomy_0.2.gpickle" 
G = nx.read_gpickle(path_taxonomy)
for n in G.nodes():
    labels[n] = n
plt.figure(figsize=(10,10))
pos = graphviz_layout(G, prog="neato")
nx.draw(G, pos, with_labels=False, arrowsize=2, arrows=False)
nx.draw_networkx_labels(G,pos,labels,font_size=18, font_color='black', font_weight='bold')
plt.show()
```
